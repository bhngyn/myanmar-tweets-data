```python
#import libraries
from datetime import date, timedelta
import twint
import pandas as pd
import nest_asyncio

#you need this for twint to work in Jupyter
nest_asyncio.apply()
```

```python
#configure the Twint search
c = twint.Config()

#configure individual componnents of the search for keywords within #MarXXCoup hashtags with a minimum of 10 likes and has media
c.Search = "body OR bodies OR corpse AND #Apr10Coup OR #Apr11Coup OR #Ar12Coup OR #WhatsHappeningInMyanmar OR #Myanmar AND since:2021-04-10"
c.Media = True
c.Min_likes = 10
c.Pandas = True

#run the search
twint.run.Search(c)
```

```python
#put the collected data into a pandas dataframe called "tweets_df"
tweets_df = twint.storage.panda.Tweets_df

#show what's inside the dataframe
tweets_df.info()
```
